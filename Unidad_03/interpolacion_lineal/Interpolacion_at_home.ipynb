```python
import numpy as np
import matplotlib.pyplot as plt

datos = np.loadtxt("datos.dat")
x_list=[]
y_list=[]

for i in range(0,datos.shape[0]):
    x_list.append(datos[i][0])
    y_list.append(datos[i][1])
    

def InterpoladorLineal(x_list,y_list,x):
    for i in range(0,len(x_list)-1):
        if (x>=x_list[i] and x<=x_list[i+1]):
            x_prev = x_list[i]
            y_prev = y_list[i]
            x_next = x_list[i+1]
            y_next = y_list[i+1]
            
            y = (y_next - y_prev)/(x_next - x_prev)*(x-x_prev) + y_prev

    return y
            
x_interpolado=np.linspace(min(x_list),max(x_list),1000)
y_interpolado=[]
for i in range(0,len(x_interpolado)):
    y_interpolado.append(InterpoladorLineal(x_list,y_list,x_interpolado[i]))
    

    
x=np.linspace(0,2*np.pi,1000)
y=np.sin(x)

fig=plt.figure()
ax=plt.axes()
ax.plot(x_list,y_list,'*')
ax.plot(x_interpolado,y_interpolado,'.')
ax.plot(x,y)



#print(InterpoladorLineal(x_list,y_list,2))
            
            
            
```
