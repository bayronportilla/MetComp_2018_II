```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,2*np.pi,1000)
y=np.sin(x)

N=8
dx = 2.0*np.pi/N
x_reduced=[]
y_reduced=[]

for i in range(0,N):
    x_reduced.append(i*dx)
    y_reduced.append(np.sin(i*dx))

fig=plt.figure()
ax=plt.axes()
ax.plot(x,y)
ax.plot(x_reduced,y_reduced,'*')

file=("datos.dat","w")
for i in range(0,len(x_reduced)):
    file.write("%.6f\t %.6f"%(x_reduced[i],y_reduced[i]))

file.close()
```
